```python
# 패키지, 모듈 가져오기
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
# 스타벅스 사이트 연결
driver = wd.Chrome()
driver.get('https://www.starbucks.co.kr/store/store_map.do')
```

```python
# 지역 검색 클릭
driver.find_element(By.CSS_SELECTOR, '.loca_search > h3 > a').click()
time.sleep(3)

# 서울 클릭
driver.find_element(By.CSS_SELECTOR, '.sido_arae_box > li:nth-child(1) > a').click()
time.sleep(3)

# 전체 클릭
driver.find_element(By.CSS_SELECTOR, '#mCSB_2_container > ul > li:nth-child(1) > a').click()
```

```python
# 서울 지역 스타벅스 정보 리스트로 담기
sb_seoul_list = driver.find_elements(By.CSS_SELECTOR, '#mCSB_3_container > ul > li')
# len(sb_seoul_list)
```

```python
# 각 매장별 정보 하나씩 뽑아서 딕셔너리 형태로
sb_seoul = [
    {
        "매장명": sb.get_attribute('data-name'),
        "주소": sb.get_attribute('textContent').split('\t')[2][:-9],
        "매장 타입": sb.find_element(By.CSS_SELECTOR,'i').get_attribute('class')[4:]
    }
        for sb in sb_seoul_list
]
# print(sb_seoul)
```

```python
# csv로 저장
final_sb = pd.DataFrame(sb_seoul)
final_sb.to_csv('송아름.csv', index=False, encoding='utf-8')
```

```python
# 웹 종료
driver.quit()
```
